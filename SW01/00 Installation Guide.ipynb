# AISO Python Installation Guide

## Install Python

* Download and install [Python](https://www.python.org/downloads/) version 3 (3.7 is recommended)
* Linux users can use their favorite package manager
* Mac users can use [Homebrew](https://brew.sh/)
* Alternatively you can use [Anaconda](https://www.anaconda.com/)

When installing python, make sure that you install **pip** and **add python to your environment variables**:


### Install Packages

For the exercises the following packages are required:
* jupyter notebook
* numpy
* pandas
* scikit-learn
* matplotlib
* tqdm
* networkx
* ortools

You can install or upgrade them with the following command using pip:
* `pip3 install --upgrade jupyter numpy pandas scikit-learn matplotlib tqdm networkx ortools`

## Jupyter Notebook

After you have installed all the required packages you can launch a jupyter notebook:

* Execute the command `jupyter notebook` within the course folder


### Most important shortcuts

* `Esc`: go to command mode
* `Enter`: go to edit mode
* `m`: switch cell to markdown
* `y`: switch cell to code
* `Shift-Enter`: Execute cell
* `Alt-Enter`: Execute cell and create new
* `dd`: delete cell
* `a`: Insert new cell above
* `b`: Insert new cell below
* `k`: Move to cell above
* `j`: Move to cell below
